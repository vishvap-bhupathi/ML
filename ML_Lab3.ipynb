```python
import pandas as pd
import numpy as np
missing_values = ["n/a", "na", "--","?"]
df = pd.read_excel("embeddingsdata.xlsx",na_values = missing_values)
df
```

```python
class1 = df[df['Label'] == 0]
class2 = df[df['Label'] == 1]
class1
```

```python
import numpy as np
mean1 = np.mean(class1)
mean2 = np.mean(class2)

variance1 = np.var(class1)
variance2 = np.var(class2)

std1=np.std(class1)
std2=np.std(class2)

interclass_spread = variance1 + variance2
interclass_distance = np.abs(mean1 - mean2)

print("InterClass Spread: ",interclass_spread)
print("InterClass Distance: ",interclass_distance)
print("Mean of Class 1:", mean1)
print("Spread of Class 1:", std1)
print("Mean of Class 2:", mean2)
print("Spread of Class 2:", std2)
```

```python
distance=np.linalg.norm(mean1-mean2)
print("Euclidean distance: ", distance)
```
