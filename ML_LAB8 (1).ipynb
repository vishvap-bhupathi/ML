```python
import pandas as pd
import numpy as np
df=pd.read_excel('lab3_data.xlsx')
df# Count occurrences of buys_computer = 'yes' and 'no'
```

```python
df = pd.DataFrame(df)
buys_computer_counts = df['buys_computer'].value_counts()
total_instances = len(df)
prior_probability_yes = buys_computer_counts['yes'] / total_instances
prior_probability_no = buys_computer_counts['no'] / total_instances
print("Prior probability of buys_computer = yes:", prior_probability_yes)
print("Prior probability of buys_computer = no:", prior_probability_no)
```

```python
from sklearn.naive_bayes import GaussianNB
df_encoded = pd.get_dummies(df.drop(columns=['buys_computer']))
gnb = GaussianNB()
gnb.fit(df_encoded, df['buys_computer'])
class_conditional_densities = gnb.theta_
print("Class-conditional densities:")
print(class_conditional_densities)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
label_encoders = {}
for column in df.columns[:-1]:  
    label_encoders[column] = LabelEncoder()
    df[column] = label_encoders[column].fit_transform(df[column])
df['age'] = df['age'].replace({'31…40': '31-40'})
X = df.drop('buys_computer', axis=1)
y = df['buys_computer']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of Gaussian Naive Bayes classifier: ",accuracy)
```
