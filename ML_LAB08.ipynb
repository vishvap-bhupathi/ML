```python
import pandas as pd
import numpy as np
missing_values = ["n/a", "na", "--","?"]
df = pd.read_excel("lab3_data.xlsx",na_values = missing_values)
df
```

```python
# Calculating prior probabilities for 'buys_computer' classes
prior_probabilities=df['buys_computer'].value_counts(normalize=True)
print("Prior Probabilities for each class:")
print(prior_probabilities)
```

```python
# Function to calculate class-conditional densities for a feature given a class
def calculate_class_conditional_density(feature,target):
    class_conditional_density = df[df['buys_computer']==target][feature].value_counts(normalize=True)
    return class_conditional_density

# Calculate class-conditional densities for each feature and class combination
features=['age','income','student','credit_rating']
classes=df['buys_computer'].unique()

for feature in features:
    print(f"Class-conditional densities for '{feature}':")
    for target_class in classes:
        density = calculate_class_conditional_density(feature, target_class)
        print(f"Class '{target_class}':")
        print(density)
        print()
```
