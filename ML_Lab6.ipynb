```python
import pandas as pd
import numpy as np
missing_values = ["n/a", "na", "--","?"]
df = pd.read_excel("embeddingsdata.xlsx",na_values = missing_values)
df
```

```python
import seaborn as sns
#plotting a scatter plot
sns.relplot(data=df,x="embed_121",y="embed_342",kind="scatter")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

#taking the independent variable as X and dependent varaible as Y 
X=df['embed_474']
Y=df['Label']
X=X.values.reshape(-1, 1)
#splitting the data into train and test data
X_train=X[:-20]
X_test=X[-20:]
Y_train=Y[:-20]
Y_test=Y[-20:]
#developing a linear regression model
regr = linear_model.LinearRegression()
#training the model using train data
regr.fit(X_train,Y_train)
#making predictions using test data
Y_pred = regr.predict(X_test)
#calculating the mean squared error
print("Mean squared error: %.2f" % mean_squared_error(Y_test, Y_pred))

```

```python
#plotting the output for linear regression model
plt.scatter(X_test, Y_test, color="black")
plt.plot(X_test,Y_pred, color="blue", linewidth=3)
plt.xticks(())
plt.yticks(())
plt.show()
```
